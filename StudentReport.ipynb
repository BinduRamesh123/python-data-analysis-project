```python
# Student Report Card System

# Function to calculate grade based on average marks
def calculate_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 75:
        return "B"
    elif avg >= 50:
        return "C"
    else:
        return "Fail"

# Function to add a new student
def add_student(students):
    roll = input("Enter Roll Number: ")
    name = input("Enter Student Name: ")
    subjects = {}
    num_subjects = int(input("Enter number of subjects: "))

    for i in range(num_subjects):
        subject = input(f"Enter subject {i+1} name: ")
        marks = int(input(f"Enter marks for {subject}: "))
        subjects[subject] = marks

    students[roll] = {"name": name, "marks": subjects}
    print("Student added successfully!\n")
# Function to generate report card
def generate_report(students):
    roll = input("Enter Roll Number to view report: ")
    if roll in students:
        info = students[roll]
        marks = info["marks"].values()
        avg = sum(marks) / len(marks)
        grade = calculate_grade(avg)

        print("\n--- Report Card ---")
        print(f"Roll Number: {roll}")
        print(f"Name: {info['name']}")
        for subject, mark in info["marks"].items():
            print(f"{subject}: {mark}")
        print(f"Average: {avg:.2f}")
        print(f"Grade: {grade}")
        print("-------------------\n")
    else:
        print("Student not found!\n")

# Function to view class statistics
def class_statistics(students):
    if not students:
        print("No student records available.\n")
        return
    all_marks = []
    for info in students.values():
        all_marks.extend(info["marks"].values())

    highest = max(all_marks)
    lowest = min(all_marks)
    avg_class = sum(all_marks) / len(all_marks)

    print("\n--- Class Statistics ---")
    print(f"Highest Marks: {highest}")
    print(f"Lowest Marks: {lowest}")
    print(f"Class Average: {avg_class:.2f}")
    print("------------------------\n")

# Main Program
students = {}

while True:
    print("1. Add Student")
    print("2. Generate Report Card")
    print("3. View Class Statistics")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_student(students)
    elif choice == "2":
        generate_report(students)
    elif choice == "3":
        class_statistics(students)
    elif choice == "4":
        print("Exiting program. Goodbye!")
        break
    else:
        print("Invalid choice! Please try again.\n")
```
